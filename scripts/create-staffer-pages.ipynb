```python
import os
import re
import csv
import requests
from PIL import Image
from io import BytesIO

def download_and_save_images(csv_content, directory):
    # Create directory if it doesn't exist
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    # Parse the CSV content line by line
    reader = csv.reader(csv_content.splitlines())
    #next(reader)  # Skip the header row
    
    for row in reader:
        print(row)
        timestamp, full_name, pronouns, link = row[0].split("\t")
        first_name, last_name = full_name.split()
        
        # Format the file name
        file_name = f"{last_name.lower()}-{first_name.lower()}.jpg"
        file_path = os.path.join(directory, file_name)
        
        match = re.search(r'/d/([^/]+)/', link)
        if match:
            file_id = match.group(1)
            download_link = f"https://drive.google.com/uc?export=download&id={file_id}"
        else:
            print(f"Failed to extract file ID from link for {full_name}. Skipping.")
            continue
        
        try:
            # Download the image using the modified download link
            response = requests.get(download_link)
            response.raise_for_status()
            
            # Open the image and convert it to JPEG if necessary
            img = Image.open(BytesIO(response.content))
            if img.format != 'JPEG':
                img = img.convert('RGB')
            
            # Save the image to the specified directory
            img.save(file_path, 'JPEG')
            print(f"Image for {full_name} has been saved successfully to {file_path}")
        
        except Exception as e:
            print(f"Failed to download and save image for {full_name}. Error: {e}")

# Copy-paste the first four columns from the staff onboarding form responses
csv_content = """9/19/2024 15:32:10	Aathi Muthu	he/him	https://drive.google.com/file/d/1Jta_xgahg1l4JvLQVaS9ME8fnARkjL-v/view?usp=sharing
9/19/2024 9:34:20	Ashley Ho	she/her	https://drive.google.com/file/d/1J1CjsEvzpwtIXh_9Qs13T-LgPu8vqxBJ/view?usp=drive_link
9/22/2024 21:35:09	Athulith Paraselli	he/him	https://drive.google.com/file/d/1ZLEx-0pdfsjVfIbvNqkDL43Z93G3KuCj/view?usp=drivesdk
9/22/2024 12:32:31	Bill Wang	he/him	https://drive.google.com/file/d/1SQNKL8hUxdpwIXB9cPNT6lqktqaWxhB8/view?usp=sharing
9/20/2024 16:31:43	Charlie Gillet	he/him	https://drive.google.com/file/d/1DTmypev9YYWUnpocoHmfv1IoXYV9E2iS/view?usp=drive_link
9/20/2024 18:55:41	Ciro Zhang	he/him	https://drive.google.com/file/d/1sw0lE791jZmE3fRRtuHlT9a8ibpZ2pWi/view?usp=sharing
9/19/2024 10:05:58	Claire Wang	she/her	https://drive.google.com/file/d/1wZ4u0cmXuRu_0hBRLYsTlsrttk-KGo2F/view?usp=sharing
9/20/2024 0:37:03	Daniel Budidharma	he/him	https://drive.google.com/file/d/1xwsVBFTvo6KaiobE2BEn9KSWhhn9SXnS/view?usp=sharing
9/19/2024 12:13:12	Dhruv Sehgal	he/him	https://drive.google.com/file/d/17EwLIyA9xByeJA_C6C8WDDvSSG09K-sg/view?usp=sharing
9/20/2024 21:43:04	Eric Chen	he/him	https://drive.google.com/file/d/1VcAgsnpBEIEFev2k3j1veITc4rWms4Jg/view?usp=sharing
9/19/2024 12:18:36	Francisco Franco 	he/him	https://drive.google.com/file/d/1b8m3p5FCeHMDM4RIcjWuR1kF0PdICzzL/view?usp=drive_link
9/19/2024 9:27:52	Hargen Zheng	he/him	https://drive.google.com/file/d/15_oMfo7QbNsFyTJaal0zxR_TgOZilwlk/view?usp=sharing
9/19/2024 9:56:37	Jack Determan	he/him	https://drive.google.com/file/d/1tEa2_-V4HhNu_TYwxwUQFg2AP-PXd2Um/view?usp=drive_link
9/19/2024 11:12:41	Jason Huynh	he/him	https://drive.google.com/file/d/1_wrHK-odM3sSvpySyulZcAGyc0jVN1eE/view?usp=sharing
9/19/2024 15:36:49	Kate Feng	she/her	https://drive.google.com/file/d/1j1t2dPwPL-AFVdFJJhJx9M1eNl2-xZHl/view?usp=sharing
9/19/2024 11:32:05	Kathleen Nguyen	she/her	https://drive.google.com/file/d/1SaO7j_5wPSYHVyE_KMbTXKDRPBjigMfK/view?usp=sharing
9/19/2024 7:02:12	Matilda Gaddi	she/her	https://drive.google.com/file/d/1SCencJ5l57SYpl4RA2Y53mezoLX3Kcwt/view?usp=drive_link
9/20/2024 0:51:06	Michelle Hong	she/her	https://drive.google.com/file/d/1I5S2U81eUyyzvbFOw9_Fro06U-1yf_ih/view?usp=sharing
9/19/2024 6:37:11	Minchan Kim	he/him	https://drive.google.com/file/d/1qpfrHjpjwGOPSnQmBbe1yFnCQAbntzqo/view?usp=sharing
9/20/2024 20:14:37	Pallavi Prabhu	she/her	https://drive.google.com/file/d/1ohZshbfRa9V2A-tACoXrtSYaCTc-oWX6/view?usp=sharing
9/19/2024 8:23:18	Pranav Rajaram	he/him	https://dsc-courses.github.io/dsc10-2024-sp/assets/staff-images/sp24/rajaram-pranav.jpg
9/19/2024 11:03:16	Raine Hoang	they/them	https://drive.google.com/file/d/1-Lxj8yY1REXiJym1caTnncbCFYi4O5RR/view?usp=sharing
9/20/2024 12:39:05	Sofia Tkachenko	she/her	https://drive.google.com/file/d/12_B5BDtQCgSzToeFfoygW_WedO_cx5sn/view?usp=drivesdk"""

# The directory where the images should be saved
directory = "../assets/staff-images/sp24"

# Call the function with the CSV content and the directory
download_and_save_images(csv_content, directory)

# This will extract some, others will need to be done manually.
```
